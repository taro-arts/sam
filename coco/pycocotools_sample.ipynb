### 精度評価

結果ファイルのフォーマット
https://cocodataset.org/#format-results

### 評価用関数の定義

````python
from pycocotools.cocoeval import COCOeval

def evaluate(coco_gt, result_file: str, iou_type: str):
    coco_dt = coco_gt.loadRes(resFile=result_file)
    
    coco_eval = COCOeval(cocoGt=coco, cocoDt=coco_dt, iouType=iou_type)
    coco_eval.evaluate()
    coco_eval.accumulate()
    coco_eval.summarize()

def evaluate_segmentation(coco_gt, result_file: str):
    evaluate(coco_gt, result_file, iou_type="segm")

def evaluate_bbox(coco_gt, result_file: str):
    evaluate(coco_gt, result_file, iou_type="bbox")

def evaluate_keypoints(coco_gt, result_file: str):
    evaluate(coco_gt, result_file, iou_type="keypoints")
````

##### 1. 物体検出
   - バウンディングボックス
        ```
        [{
            "image_id": int,
            "category_id": int, 
            "bbox": [x,y,width,height], 
            "score": float,
        }]
        ```

````python
import random
import json
from pathlib import Path

random.seed(0)

def make_bbox_detection_result_file(filepath, coco, is_noise=False):
    results = []
    for ann_id, ann in coco.anns.items():
        bbox = ann["bbox"].copy()
        
        # 正解にわざとズレを加えてダミーの結果を生成
        if is_noise:
            bbox[0] += random.random()
            bbox[1] += random.random()
            bbox[2] += random.random()
            bbox[3] += random.random()
        
        result = {
            "image_id": ann['image_id'],
            "category_id": ann["category_id"],
            "bbox": bbox,
            "score": 1.0,
        }
        results.append(result)

    Path(filepath).parent.mkdir(parents=True, exist_ok=True)
    with open(filepath, "w") as fp:
        json.dump(results, fp, indent=None)

# 正解データ
annotation_file = r'.annotations\instances_val2017.json'
coco_gt = pycocotools.coco.COCO(annotation_file)

# 結果ファイル作成
bbox_detection_result_file = ".result/bbox_detection.json"
make_bbox_detection_result_file(bbox_detection_result_file, coco_gt, is_noise=True)

# 評価
evaluate_bbox(coco_gt, bbox_detection_result_file)
````

##### 1. 物体検出
   - セグメンテーション
        ```
        [{
            "image_id": int, 
            "category_id": int, 
            "segmentation": RLE, 
            "score": float,
        }]
        ```


````python
import random
import json
from pathlib import Path

random.seed(0)

def make_segm_detection(filepath, coco, is_noise=False):
    results = []
    for ann_id, ann in coco.anns.items():
        ann_ = ann.copy()        
        segm = ann_["segmentation"].copy()
        
        # 正解にわざとズレを加えてダミーの結果を生成
        if is_noise:
            raise NotImplementedError()
    
        # RLE に変換
        ann_["segmentation"] = segm
        segm_rle = coco.annToRLE(ann_)
        segm_rle["counts"] = segm_rle["counts"].decode('utf-8')
        # print(segm_rle)
        
        result = {
            "image_id": ann_['image_id'],
            "category_id": ann_["category_id"],
            "segmentation": segm_rle,
            "score": 1.0,
        }
        results.append(result)

    Path(filepath).parent.mkdir(parents=True, exist_ok=True)
    with open(filepath, "w") as fp:
        json.dump(results, fp, indent=None)

# 正解データ
annotation_file = r'.annotations\instances_val2017.json'
coco_gt = pycocotools.coco.COCO(annotation_file)

# 結果ファイル作成
segm_detection_result_file = ".result/segm_detection.json"
make_segm_detection(segm_detection_result_file, coco_gt, is_noise=False)

# 評価
evaluate_segmentation(coco_gt, segm_detection_result_file)
````
